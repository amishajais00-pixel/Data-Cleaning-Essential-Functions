1. What is data cleaning, and why is it important in data analysis?
What are the potential consequences of analyzing unclean or messy data?
Explain the common steps involved in cleaning and organizing data                                                          
Data cleaning (or data cleansing) is the process of correcting, removing, or transforming inaccurate, incomplete, or inconsistent data to ensure reliability. It’s essential because unclean data can lead to misleading insights, poor decisions, and faulty models.
Common steps include handling missing values, removing duplicates, correcting errors, standardizing formats, and dealing with outliers.

5. Apply Data Validation to restrict Quantity values to only whole numbers between 1 and 10.
Configure an input message that appears when a user selects a cell in the "Quantity" column explaining:


a. **Select the Quantity cells**
   - Choose the cells under the “Quantity” column where users will enter numbers (e.g., all rows in that column).

b. **Open Data Validation**
   - Go to the Data tab.
   - Click **Data Validation** → **Data Validation…**.

c. **Set the validation rule**
   - In the **Settings** tab:
     - **Allow:** Whole number  
     - **Data:** Between  
     - **Minimum:** 1  
     - **Maximum:** 10


d. **Add a helpful message on selection**
   - Switch to the **Input Message** tab.
   - Check **Show input message when cell is selected**.
   - **Title:** Quantity rule
   - **Input message:** Please enter a whole number between 1 and 10.


e. **Create a clear error alert**
   - Go to the **Error Alert** tab.
   - Check **Show error alert after invalid data is entered**.
   - **Style:** Stop (to block invalid entries)
   - **Title:** Invalid input
   - **Error message:** Invalid input! The quantity must be a whole number between 1 and 10.

f. **Test the validation**
   - Click **OK** to apply.
   - Try entering 0, 3.5, 11, and 7 to confirm:
     - **0 / 11 / 3.5:** Triggers the error alert.
     - **7:** Accepted.
